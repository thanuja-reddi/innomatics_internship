```python
# 1. Smart light controller
class LightController:
    def __init__(self,name):
        self.name=name
        self.action="OFF"
    def turn_on(self):
        self.action="ON"
    def turn_off(self):
        self.action="OFF"
    def display_result(self):
        print(self.name,"is",self.action)
light=lightcontroller("Bedroom Light")
light.turn_on()
light.display_result()
```

```python
# 2. Employee ID card system
class Employee:
    def __init__(self,name,id,dept):
        self.name=name
        self.id=id
        self.dept=dept
    def display_result(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.id)
        print("Department:",self.dept)
emp=Employee("Rahul","EMP102","IT")
emp.display_result()
```

```python
# 3. Mobile contact record
class Mobile:
    def __init__(self,name,phoneno):
        self.name=name
        self.phoneno=phoneno
    def display_result(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone:",self.phoneno)
mob=Mobile("Anita",'9876543210')
mob.display_result()   
```

```python
# 4. Product price tag generator
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display_result(self):
        print("Product:",self.name)
        print("Price:$",self.price)
price=Product("Headphones",2499)
price.display_result()
```

```python
# 5. Movie rating display system
class MovieRating:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display_result(self):
        print("Movie Name:",self.name)
        print("Rating:",self.rating,"/5")
rating=MovieRating("Inception",4.8)
rating.display_result()
```

```python
# 6. Delivery Address manager
class DeliveryAddress:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def display_result(self):
        print("Delivery Details")
        print("Customer:",self.name)
        print("Address:",self.address)
address=DeliveryAddress("Suman","Hyderabad")
address.display_result()
```
