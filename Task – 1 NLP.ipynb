```python
# Import Libraries
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()


class NLPPreprocessor:

    def __init__(self):
        self.stop_words = stop_words
        self.stemmer = stemmer
        self.lemmatizer = lemmatizer

    # 1. Lowercase
    def to_lowercase(self, text):
        return text.lower()

    # 2. Remove URLs
    def remove_urls(self, text):
        return re.sub(r'http\S+|www\S+|https\S+', '', text)

    # 3. Remove HTML Tags
    def remove_html(self, text):
        return re.sub(r'<.*?>', '', text)

    # 4. Remove Emojis
    def remove_emojis(self, text):
        emoji_pattern = re.compile("["
                                   u"\U0001F600-\U0001F64F"
                                   u"\U0001F300-\U0001F5FF"
                                   u"\U0001F680-\U0001F6FF"
                                   u"\U0001F1E0-\U0001F1FF"
                                   "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)

    # 5. Remove Punctuation
    def remove_punctuation(self, text):
        return text.translate(str.maketrans('', '', string.punctuation))

    # 6. Remove Numbers
    def remove_numbers(self, text):
        return re.sub(r'\d+', '', text)

    # 7. Tokenization
    def tokenize(self, text):
        return word_tokenize(text)

    # 8. Remove Stopwords
    def remove_stopwords(self, tokens):
        return [word for word in tokens if word not in self.stop_words]

    # 9. Stemming
    def apply_stemming(self, tokens):
        return [self.stemmer.stem(word) for word in tokens]

    # 10. Lemmatization
    def apply_lemmatization(self, tokens):
        return [self.lemmatizer.lemmatize(word) for word in tokens]

    # 11. Full Pipeline
    def preprocess(self, text):

        print("Original Text:\n", text)

        text = self.to_lowercase(text)
        text = self.remove_urls(text)
        text = self.remove_html(text)
        text = self.remove_emojis(text)
        text = self.remove_punctuation(text)
        text = self.remove_numbers(text)

        tokens = self.tokenize(text)
        tokens = self.remove_stopwords(tokens)

        stemmed = self.apply_stemming(tokens)
        lemmatized = self.apply_lemmatization(stemmed)

        print("\nProcessed Tokens:\n", lemmatized)

        return lemmatized


# ================================
# Example Usage
# ================================

if __name__ == "__main__":

    preprocessor = NLPPreprocessor()

    sample_text = """
    Hello!!! 😊 This is an NLP Preprocessing Example.
    Visit https://example.com for more details.
    We are learning Data Science in 2026 🚀.
    """

    result = preprocessor.preprocess(sample_text)

texts = [
    "Machine Learning is AMAZING!!! 🤖",
    "I scored 95% in exams!!! 📚",
    "<p>This is HTML content</p>",
    "Visit www.google.com now!"
]

print("\n\n--- Additional Examples ---")

for t in texts:
    print("\nInput:", t)
    print("Output:", NLPPreprocessor().preprocess(t))
```
