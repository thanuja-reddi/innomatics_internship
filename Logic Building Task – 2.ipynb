```python
# 1. Unique words in a sentence
sentence=input("enter sentence:")
words=sentence.split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

```python
# 2. Highest Salary from employee data
employees={"Ravi":75000,"Anita":68000,"Kiran":72000}
max_val=0
employee=""
for name in employees:
    if(employees[name]>max_val):
        max_val=employees[name]
        employee=name
print("Highest Salary:",employee,"-",max_val)
        
    
```

```python
# 3. Find maximum and minimum vales
numbers=[45,22,89,10,66]
min_val=numbers[1]
max_val=numbers[1]
for num in numbers:
    if(num<min_val):
        min_val=num
    if(num>max_val):
        max_val=num
print("List:",numbers)
print("Max:",max_val)
print("Min:",min_val)
```

```python
# 4. Count products above a price threshold
prices=[450,1200,899,1500,300]
count=0
for price in prices:
    if(price>1000):
        count+=1
print("Products above 1000:",count)
        
```

```python
# 5. Calculate attendance percentage
attendance=["P","P","A","P","P"]
present=0
for attend in attendance:
    if(attend=="P"):
        present+=1
print("Attendance Percentage:",(present/len(attendance))*100)       
```

```python
# 6. Remove duplicate phone numbers
phone_numbers=[9876543210,9123456789,9876543210]
unique=set(phone_numbers)
print("Unique phone numbers:",unique)
```

```python
# 7. Count character frequency
text=input("enter text:")
freq={}
for char in text:
    if char in freq:
        freq[char]+=1
    else:
        freq[char]=1
print(freq)
```

```python
# 9. Chech if a key exists in a dictionary
employees={"Ravi":750000,"Anita":600000,"Sita":875000}
key=input("enter the value:")
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
