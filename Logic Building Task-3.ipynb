```python
# 1. Employee performance bonus eligibility
employees={"Ravi":92,"Anita":88,"Kiran":92,"Suresh":85}
highest_score=0
for name in employees:
    if employees[name]> highest_score:
        highest_score=employees[name]
print("Top performers eligible for bonus:",end="")
for name in employees:
    if employees[name] == highest_score:
        print(name,end=" ")
print("(Score:",highest_score,")")
```

```python
# 2. Search query keyword analysis
query="Buy mobile phone buy phone online"
query=query.lower()
words=query.split()
word_count={}
for word in words:
    if  word in word_count:
        word_count[word]+=1
    else:
        word_count[word]=1
for word in word_count:
    if word_count[word]>1:
        print(word,":",word_count[word])
        
```

```python
# 3. Sensor data validation
sensor_readings=[3,4,7,8,10,12,5]
valid_readings = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 ==0:
        pair=(i,sensor_readings[i])
        valid_readings.append(pair)
print("Valid sensor readings (hour, value):")
print(valid_readings)
```

```python
# 4. Email domain usage analysis
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_count ={}
for email in emails:
    parts=email.split("@")
    domain=parts[1]
    if domain in domain_count:
        domain_count[domain]+=1
    else:
        domain_count[domain]=1
total=len(emails)
for domain in domain_count:
    percentage=(domain_count[domain]/total)*100
    print(domain,":",int(percentage),"%")
    
```

```python
# 5. Sales spike detection
sales=[1200,1500,900,2200,1400,3000]
total=0
for value in sales:
    total+=value
average=total/len(sales)
limit=average*1.2
for i in range(len(sales)):
    if sales[i]>limit:
        print("Day",i+1,":",sales[i])
```

```python
# 6.Duolicate user id detction
user_ids=["user1","user2","user1","user3","user1","user3"]
count_ids={}
for user in user_ids:
    if user in count_ids:
        count_ids[user]+=1
    else:
        count_ids[user]=1
for user in count_ids:
    if count_ids[user]>1:
        print(user,"->",count_ids[user],"times")
```
