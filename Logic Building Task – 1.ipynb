```python
# 1.User Login Check
username="admin"
password="1234"
input_username=input()
input_password=input()
if(input_username==username and input_password==password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2.Pass/Fail Analyzer
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for mark in marks:
    if mark>=50:
        pass_count+=1
    else:
        fail_count+=1
print("Pass Count:",pass_count)
print("Fail Count:", fail_count)
```

```python
# 3.Simple Data Cleaner
names=["Alice"," bob","CHARLIE"]
cleaned_names=[]
for name in names:
    cleaned=name.strip().lower()
    cleaned_names.append(cleaned)
print("Cleaned_names",cleaned_names)
    
```

```python
# 4.Message Length Analyzer
messages=["Hi","Welcome to the platform","OK"]
for message in messages:
    length=len(message)
    if(length>=10):
        print(message,":",length,"long message")
    else:
        print(message,length)     
```

```python
# 5.Error Message Detector
logs=["INFO","ERROR","WARNING","ERROR"]
error=0
for log in logs:
    if(log=="ERROR"):
        error+=1
print("No.of ERROR entries",error)
```
