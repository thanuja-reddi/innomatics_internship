```python
# 1. Smart parking lot management system
def manage_parking(capacity,logs):
    parked=0
    peak=0
    for log in logs:
        if log=="IN":
            parked+=1
        elif log=="OUT" and parked>0:
            parked-=1
        peak=max(peak,parked)
    print("Currently Parked Vehicles:",parked)
    print("Peak Parking Usage:",peak)
    if parked>capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
capacity=int(input("Enter Parking Capacity:"))
logs=input("enter the logs").upper().split()
manage_parking(capacity,logs)
            
```

```python
# 2 Online Food Delivery time estimator
def time_estimator(distance,weather_condition,traffic_level):
    time=distance*10
    if weather_condition=="rainy":
        time+=20
    elif weather_condition=="sunny":
        time+=5
    if traffic_level=="high":
        time+=15
    elif traffic_level=="medium":
        time+=10
    print("Estimated Delivery Time:",time,"minutes")
distance=int(input("enter distance in kms:"))
weather_condition=input("enter the weather condition:").lower()
traffic_level=input("enter the traffic level:").lower()
time_estimator(distance,weather_condition,traffic_level)    
```

```python
# 3. Movie theatre seat occupancy analyzer
def seat_analyzer(total_seats,booked):
    percent=(len(booked)/total_seats)*100
    print("Ocuupancy:",percent,"%")
    if percent==100:
        print("Show Status: Housefull")
    elif percent>=75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
total_seats=int(input("enter the max no of seats"))
booked=eval(input("enter the values like [1)*90:"))
seat_analyzer(total_seats,booked)
    
```

```python
# 4. Cloud server load classification system
def load_classification(readings):
    total=0
    for reading in readings:
        total+=reading
    avg=total/len(readings)
    if(avg<50):
        status="Normal"
    elif(avg>=50 and avg<=80):
        status="Warning"
    else:
        status="Critical"
    return avg,status
readings=list(map(int,input("enter the readings:").split()))
avg,status=load_classification(readings)
print("Average CPU Load:",avg)
print("Server Status:",status)
```

```python
# 6. Online event registration capacity controller
def online_event(capacity,registrations):
    if(registrations<=capacity):
        Confirmed=registrations
        waiting=0
        status='open'
    elif(registrations>capacity):
        Confirmed=capacity
        waiting=registrations-capacity
        status='closed'
    return Confirmed,waiting,status
capacity=int(input("enter the max capacity:"))
registrations=int(input("enter the no.of registrations"))
Confirmed,waiting,status=online_event(capacity,registrations)
print("Confirmed Registrations:",Confirmed)
print("Waitlisted Users:",waiting)
print("Registration Status:",status)
```
