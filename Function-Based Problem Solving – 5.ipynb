```python
# 1. Social Media - Post Engagement Analyzer
def post_analyzer(likes):
    total=0
    for like in likes:
        total+=like
    print("Total Likes:",total)
    if total>=1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
likes=list(map(int,input("enter the likes:").split()))
post_analyzer(likes)
        
```

```python
# 2. Healthcare - Medicine Stock Alert System
def stock_alert(stocks):
    print("Medicine Stock:",stocks)
    if stocks<10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")
stocks=int(input("enter the stocks no."))
stock_alert(stocks)
```

```python
# 3. Ariculture - Rainfall Adequacy Checker
def rainfall_adequacy_checker(rainfall,required_level):
    total=0
    for rain in rainfall:
        total+=rain
    avg=total//len(rainfall)
    print("Average Rainfall:",avg)
    if avg>=required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall=list(map(int,input("enter the rainfall values").split()))
required_level=int(input("enter the required level:"))
rainfall_adequacy_checker(rainfall,required_level)
```

```python
# 4. Social Media - Duplicate Account Detection
def check_duplicates(usernames):
    if len(usernames)!=len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users=["john","alice","john","bob"]
check_duplicates(users)
```

```python
# 5. Appointment Eligibility Checker
def appointment_eligibility(age):
    print("Patient Age:",age)
    if age>=18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")
appointment_eligibility(21)
```

```python
# 6. Premium crop price filter
def premium_crops(prices):
    premium=[p for p in prices if p>2000]
    print("Premium Crops:",premium)
prices=[1500,2500,1800,3200]
premium_crops(prices)
```

```python
# 7. Application health checker
def app_health(errors):
    print("Error Count:",errors)
    if errors==0:
        print("System Status: Healthy")
    elif errors<=5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
app_health(7)
```

```python
# 8. Daily transaction limit checker
def transaction_check(amount):
    print("Transaction Amount:",amount)
    if amount<=50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")
transaction_check(60000)
```

```python
# 9. Attendance Eligibilty system
def attendance_check(attendance_list):
    present=attendance_list.count("P")
    percentage=(present/len(attendance_list))*100
    print("Attendance Percentage:",percentage)
    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance=["P","P","A","P","P"]
attendance_check(attendance)
```
