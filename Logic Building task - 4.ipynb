```python
# 1. Mobile recharge validation system
def recharge_request(amount):
    valid_plans=[199,299,399,599]
    if amount<50:
        return "recharge rejected:amount is less than 50"
    if amount not in valid_plans:
        return "recharge rejected:please choose the valid plans"
    return "recharge successful"
while True:
    amount=int(input("enter the amount"))
    result=recharge_request(amount)
    print(result)
    if result=="recharge successful":
        break
        
```

```python
# 2. Inventory reorder alert system
def reorder_alert(stock):
    for product,quantity in stock.items():
        if quantity<15:
            print(product," Reorder Alert")
        else:
            print(product," Stock OK")
stock={"pen":20,"book":40,"highlighter":4,"scale":14,"pencil":10,"eraser":5,"sharpner":15}
reorder_alert(stock)
```

```python
# 3. Student result processing system
def student_result(marks):
    total=0
    for mark in marks:
        total+=mark
    avg=total/len(marks)
    print("average:",avg)
    if avg>=50:
        print("Pass")
    else:
        print("Fail")
marks=[50,60,70,45,30,45,80,99]
student_result(marks)
```

```python
# 5. Employee attendance eligibility
def employee_attendance(attendance_list):
    p_count=0
    for attendance in attendance_list:
        if attendance=="P":
            p_count+=1
    percentage=(p_count/len(attendance_list))*100
    print("attendace percentage:",percentage)
    if percentage>=75:
        print("Eligible")
    else:
        print("Not Eligible")
attendance_list=["P","P","A","P","A","A","P","P","P"]
employee_attendance(attendance_list)
```

```python
# 6. Password strength checker
def check_password(password):
    special_chars="@#$"
    has_digit=False
    has_special=False
    if len(password)<8:
        return "weak password length is less than 8"
    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in special_chars:
            has_special=True
    if has_digit and has_special:
        return "strong password"
        
```
